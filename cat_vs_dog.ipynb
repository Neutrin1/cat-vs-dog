```python
#
cat_path = "animals/cat"
dog_path = "animals/dog"

# Function to check if a file is an image
def is_image(file_path):
    try:
        with PIL.Image.open(file_path) as img:
            return True
    except:
        return False

# Process cat images
cat_count = 0
for filename in os.listdir(cat_path):
    file_path = os.path.join(cat_path, filename)
    if os.path.isfile(file_path) and is_image(file_path) and not filename.startswith("cat."):
        new_name = "cat." + filename
        new_path = os.path.join(cat_path, new_name)
        os.rename(file_path, new_path)
        cat_images.append(new_path)
        cat_labels.append("cat")
        all_labels.append("cat")
        cat_count += 1

# Process dog images
dog_count = 0
for filename in os.listdir(dog_path):
    file_path = os.path.join(dog_path, filename)
    if os.path.isfile(file_path) and is_image(file_path) and not filename.startswith("dog."):
        new_name = "dog." + filename
        new_path = os.path.join(dog_path, new_name)
        os.rename(file_path, new_path)
        dog_images.append(new_path)
        dog_labels.append("dog")
        all_labels.append("dog")
        dog_count += 1

print(f"Processed {cat_count} cat images and {dog_count} dog images")
```

```python
import os

train_path = "animals/train"
train_cat_path = os.path.join(train_path, "cat")
train_dog_path = os.path.join(train_path, "dog")

train_images = []
train_labels = []

# Label cat images
for filename in os.listdir(train_cat_path):
    file_path = os.path.join(train_cat_path, filename)
    train_images.append(file_path)
    train_labels.append("cat")

# Label dog images
for filename in os.listdir(train_dog_path):
    file_path = os.path.join(train_dog_path, filename)
    train_images.append(file_path)
    train_labels.append("dog")

print(f"Labeled {len(train_images)} training images")

```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os
import numpy as np
from sklearn.metrics import confusion_matrix
import itertools

import torch.nn as nn
import torch.optim as optim

# 定义一个自定义数据集
class CatDogDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        """
        初始化猫狗数据集。

        参数:
            image_paths (list): 图像路径列表。
            labels (list): 标签列表 (例如, 'cat' 或 'dog')。
            transform (callable, 可选): 应用于图像的转换。默认为 None。
        """
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        """返回数据集的总大小。"""
        return len(self.image_paths)

    def __getitem__(self, idx):
        """
        检索给定索引的图像和标签。

        参数:
            idx (int): 图像的索引。

        返回值:
            tuple: 包含已转换图像 (torch.Tensor) 和标签 (int) 的元组。
                   标签 0 代表猫，1 代表狗。
        """
        image_path = self.image_paths[idx]
        image = Image.open(image_path).convert('RGB')  # 确保图像是 RGB 格式
        label = 0 if self.labels[idx] == 'cat' else 1  # 0 代表猫, 1 代表狗
        if self.transform:
            image = self.transform(image)
        return image, label
# 定义 CNN 模型
class Net(nn.Module):
    def __init__(self):
        """
        初始化 CNN 架构。
        该架构包括两个卷积层、ReLU 激活函数、最大池化层和两个全连接层。
        """
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(64 * 56 * 56, 128)  # 调整后的输入大小
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, 2)  # 2 个输出类: 猫和狗
    def forward(self, x):
        """
        定义模型的前向传播。
        参数:
            x (torch.Tensor): 输入张量。
        返回值:
            torch.Tensor: 输出张量。
        """
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(-1, 64 * 56 * 56)  # 调整后的 reshape
        x = self.relu3(self.fc1(x))
        x = self.fc2(x)
        return x

# 数据转换
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # 将图像调整为一致的大小
    transforms.ToTensor(),           # 将图像转换为张量
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) # 规范化图像
])

# 创建数据集
train_dataset = CatDogDataset(train_images, train_labels, transform=transform)

# 创建数据加载器
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# 实例化模型
net = Net()

# 定义损失函数和优化器
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)

# 使用 CUDA (GPU) 如果可用
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
net.to(device)

# 训练模型
num_epochs = 50

for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        # 将数据移动到 GPU
        images = images.to(device)
        labels = labels.to(device)

        # 梯度归零
        optimizer.zero_grad()

        # 前向传播
        outputs = net(images)
        loss = criterion(outputs, labels)

        # 反向传播和优化
        loss.backward()
        optimizer.step()

        # 打印统计信息
        if (i + 1) % 10 == 0:
            print('Epoch [{}/{}], Step [{}/{}], Loss: {:.8f}'
                  .format(epoch + 1, num_epochs, i + 1, len(train_loader), loss.item()))

print('训练完成')
# 准备测试数据
test_path = "animals/test"
cat_path = os.path.join(test_path, "cat")
dog_path = os.path.join(test_path, "dog")

test_images = []
test_labels = []

# 加载猫的测试图像
for filename in os.listdir(cat_path):
    file_path = os.path.join(cat_path, filename)
    test_images.append(file_path)
    test_labels.append("cat")

# 加载狗的测试图像
for filename in os.listdir(dog_path):
    file_path = os.path.join(dog_path, filename)
    test_images.append(file_path)
    test_labels.append("dog")

# 创建测试数据集
test_dataset = CatDogDataset(test_images, test_labels, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# 评估模型
net.eval()  # 设置为评估模式
correct = 0
total = 0
predictions = []
true_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        predictions.extend(predicted.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

print('Accuracy of the network on the test images: %d %%' % (
    100 * correct / total))

# 混淆矩阵
cm = confusion_matrix(true_labels, predictions)

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()


# 定义类别
classes = ['cat', 'dog']

# 绘制混淆矩阵
plot_confusion_matrix(cm, classes,
                          title='Confusion matrix, without normalization')

plot_confusion_matrix(cm, classes, normalize=True,
                         title='Normalized confusion matrix')
```

```python
import PIL
import random

import matplotlib.pyplot as plt

# Function to display an image with its predicted label
def display_image_prediction(image_path, model, device, classes):
    """
    Display an image with its predicted label.

    Args:
        image_path (str): Path to the image file.
        model (torch.nn.Module): Trained PyTorch model.
        device (torch.device): Device (CPU or GPU) to use for prediction.
        classes (list): List of class names.
    """
    image = PIL.Image.open(image_path).convert('RGB')
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])
    input_tensor = transform(image).unsqueeze(0).to(device)

    model.eval()
    with torch.no_grad():
        output = model(input_tensor)
        _, predicted_idx = torch.max(output, 1)
        predicted_label = classes[predicted_idx[0]]

    plt.imshow(image)
    plt.title(f'Predicted: {predicted_label}')
    plt.show()

# Example usage: Display a random test image
test_images = []
test_path = "animals/test"
cat_path = os.path.join(test_path, "cat")
dog_path = os.path.join(test_path, "dog")

# Load cat test images
for filename in os.listdir(cat_path):
    file_path = os.path.join(cat_path, filename)
    test_images.append(file_path)

# Load dog test images
for filename in os.listdir(dog_path):
    file_path = os.path.join(dog_path, filename)
    test_images.append(file_path)

random_image_path = random.choice(test_images)
display_image_prediction(random_image_path, net, device, classes)
```
